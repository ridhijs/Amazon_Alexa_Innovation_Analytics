```python
import pandas as pd
import numpy as np
```

```python
alexa = pd.read_table("amazon_alexa.tsv")
```

```python
alexa.head()
```

```python
alexa.info()
```

```python
alexa['variation'].describe()
```

## Get Dummy Variables

```python
alexa = pd.get_dummies(alexa, columns = ['variation'], drop_first = True)
```

```python
alexa.head()
```

```python
alexa['date']= pd.to_datetime(alexa['date'])
```

```python
alexa.head()
```

```python
alexa['Year'] = alexa['date'].dt.year
alexa['Month'] = alexa['date'].dt.month
alexa['Day'] = alexa['date'].dt.day

alexa = alexa.drop(['date', 'Year'], axis = 1)

alexa.head()
```

```python
from textblob import TextBlob
def sentiment_calc(text):
    try:
        return TextBlob(text).sentiment
    except:
        return TextBlob('hello').sentiment
```

```python
alexa['sentiment'] = alexa['verified_reviews'].apply(lambda text: sentiment_calc(text))
alexa[['polarity', 'subjectivity']] = alexa['sentiment'].apply(pd.Series)
alexa = alexa.drop(['sentiment'], axis = 1)
alexa.head()
```

## Average Word Length

```python
def avg_word(sentence):
  try:
    words = sentence.split()
    return (sum(len(word) for word in words)/len(words))
  except:
    return 0
```

```python
alexa['number_words'] = alexa['verified_reviews'].str.split().str.len()
alexa['number_character'] = alexa['verified_reviews'].str.len()
alexa['avg_word'] = alexa['verified_reviews'].apply(lambda x: avg_word(x))
```

```python
alexa.head()
```

## Stopwords

```python
import nltk
nltk.download("stopwords")
from nltk.corpus import stopwords
stop = stopwords.words('english')
alexa['stopwords'] = alexa['verified_reviews'].apply(lambda x: len([x for x in x.split() if x in stop]))
alexa.head()
```

## Scaler Transformation

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
columns = ['number_character','number_words', 'avg_word', 'stopwords']
for col in columns:
  alexa[[col]] = scaler.fit_transform(alexa[[col]])
alexa.head()
```

## Text Summarization

```python
from gensim.summarization import summarize
def sum_text(text):
  try:
    summed_text = summarize(text)
    return summed_text
  except:
    return text
```

```python
alexa['summed_text'] = alexa['verified_reviews'].apply(lambda x: sum_text(x))
alexa['sentiment_sum'] = alexa['summed_text'].apply(lambda text: sentiment_calc(text))
alexa[['polarity_sum', 'subjectivity_sum']] = alexa['sentiment_sum'].apply(pd.Series)
alexa = alexa.drop(['sentiment_sum'], axis = 1)
alexa.head()
```

```python
alexa2 = alexa.drop(['verified_reviews', 'summed_text'], axis = 1)
alexa2.head()
```

## Visualization

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
sns_plot = sns.pairplot(alexa, hue = 'rating', vars=['polarity','subjectivity', 'number_words', 'number_character'])
```

```python
sns_plot = sns.pairplot(alexa2, hue = 'rating', vars=['polarity', 'subjectivity', 'subjectivity_sum', 'polarity_sum'])
```

## Model Building

```python
from sklearn.model_selection import train_test_split
import numpy as np
alexa1=np.matrix(alexa2.drop(['rating'], axis = 1))
y=np.array(alexa2['rating'])
X_train, X_test, y_train, y_test = train_test_split(alexa1, y, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=500, criterion='gini', max_depth=None, min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_features='auto', max_leaf_nodes=None, min_impurity_decrease=0.0, min_impurity_split=None, bootstrap=True, oob_score=False, n_jobs=None, random_state=None, verbose=0, warm_start=False, class_weight=None)
rfc.fit(X_train,y_train)
score = rfc.score(X_test,y_test)
print(score)
```

## Visualize Results

```python
from sklearn.metrics import confusion_matrix
predictions = rfc.predict(X_test)
cm = confusion_matrix(y_test, predictions)
labels = ['1','2','3','4','5']
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(cm)
plt.title('Confusion matrix of the classifier')
fig.colorbar(cax)
ax.set_xticklabels([''] + labels)
ax.set_yticklabels([''] + labels)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```
